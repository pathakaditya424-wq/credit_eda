```python



# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Read datasets
application = pd.read_csv('/content/application_data (1).csv')   # Current application dataset
previous_application = pd.read_csv('/content/previous_application (1).csv')   # Previous application dataset

# Display dataset shapes
print("Application dataset shape:", application.shape)
print("Previous application dataset shape:", previous_application.shape)
```

```python
# Step 3: Basic inspection of the data
print("\nFirst 5 rows of application data:")
display(application.head())

print("\nDataset Info:")
application.info()

print("\nSummary statistics of numerical columns:")
display(application.describe())
```

```python
# Step 4: Check for missing values
missing_values = application.isnull().sum()
missing_percentage = (missing_values / len(application)) * 100

missing_data = pd.DataFrame({
    'Missing Values': missing_values,
    'Missing %': missing_percentage
}).sort_values(by='Missing %', ascending=False)

print("Columns with missing values:")
display(missing_data[missing_data['Missing Values'] > 0].head(15))  # Top 15 missing
```

```python
# Step 5: Handle missing values
# Numeric columns → replace with median
num_cols = application.select_dtypes(include=np.number).columns
application[num_cols] = application[num_cols].fillna(application[num_cols].median())

# Categorical columns → replace with mode
cat_cols = application.select_dtypes(include='object').columns
application[cat_cols] = application[cat_cols].fillna(application[cat_cols].mode().iloc[0])

# Verify
print("Missing values after handling:", application.isnull().sum().sum())
```

```python
# Step 6: Data types check
print("\nData types of columns:")
print(application.dtypes.head(15))
```

```python
# Step 7a: Target variable distribution
sns.countplot(x='TARGET', data=application, palette='Set2')
plt.title("Loan Repayment (0 = Repaid, 1 = Default)")
plt.show()

print(application['TARGET'].value_counts(normalize=True)*100)
```

```python
# Step 7b: Distribution of key numeric features
plt.figure(figsize=(12,5))
sns.histplot(application['AMT_INCOME_TOTAL'], bins=50, kde=True)
plt.title("Applicant Income Distribution")
plt.show()

plt.figure(figsize=(12,5))
sns.histplot(application['AMT_CREDIT'], bins=50, kde=True)
plt.title("Credit Amount Distribution")
plt.show()
```

```python
# Step 7c: Analysis of gender vs target
sns.countplot(x='CODE_GENDER', hue='TARGET', data=application)
plt.title("Default Rate by Gender")
plt.show()
```

```python
# Step 8: Correlation heatmap with top features
corr = application.select_dtypes(include=['number']).corr()

# Select top 15 features most correlated with TARGET
top_corr = corr['TARGET'].abs().sort_values(ascending=False).head(15).index

plt.figure(figsize=(10,6))
sns.heatmap(corr.loc[top_corr, ['TARGET']], annot=True, cmap="coolwarm", center=0)
plt.title("Top Correlated Features with TARGET", fontsize=14)
plt.show()
```
